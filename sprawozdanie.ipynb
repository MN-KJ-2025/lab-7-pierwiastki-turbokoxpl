<br>

**Wielomian**

Dany jest wielomian $w(x)$ w postaci kanonicznej (ogólnej):

$$
w(x) = a_nx^n + a_{n-1}x^{n-1} + \ldots + a_2x^2 + a_1x + a_0
$$

jego wektor współczynników wyraża się wzorem:

$$
\mathbf{a} = (a_0, a_1, ..., a_{n-1}, a_n)
$$

**Punkt 3.**

Zestawienie wyników.

1. Utwórz wykres typu *scatter plot* i wyrysuj na nim wartości pierwiastków wyznaczonych w **Punkcie 2.** oraz pierwiastki oblicznone analitycznie na płaszczyźnie zespolonej.

2. Opisz wnioski wynikające z wykresu.

````python
# ===== Twoja implementacja tutaj =====
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as nppoly
import main
w3 = nppoly.polyfromroots([1]*8)

Fw3 = main.frob_a(w3)
print("Macierz Frobeniusa F_w3:\n", Fw3)


własne = np.linalg.eigvals(Fw3)
print("\nWartości własne F_w3:\n", własne)


T,Z=sp.linalg.schur(Fw3)
schur = np.diag(T)
print("\nWartości własne z rozkładu Schura:\n", schur)



pierwiastki = nppoly.polyroots(w3)
print("\nPierwiastki polyroots:\n", pierwiastki)

analitycznie = np.array([1]*8, dtype=float)


plt.scatter(pierwiastki.real, pierwiastki.imag, s=60, label="nppoly.polyroots")
plt.scatter(analitycznie.real, analitycznie.imag, s=60, label="wartości analityczne")

plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(True)
plt.xlabel("Re")
plt.ylabel("Im")
plt.title("Pierwiastki wielomianu w3(x) = (x-1)^8")
plt.legend()
plt.show()
````
